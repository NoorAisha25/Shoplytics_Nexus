<h2 style="color:skyblue; font-size:28px;">🔵 Gender :</h2>


```python
#plotting a pie chart for gender distribution

gender_counts = df['Gender'].value_counts()

labels = gender_counts.index
sizes = gender_counts.values
colors = ['skyblue', 'green']  

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Gender Distribution',fontweight='bold',fontsize='16')
plt.legend(wedges,
    labels,
    title='Gender',
    loc='center left',           
    bbox_to_anchor=(1, 0.5),    
    fontsize=10
)
plt.axis('equal')  
plt.show()

```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

labels = sales_gen['Gender']
sizes = sales_gen['Amount']
colors = ['lightcoral', 'lightskyblue'] 

plt.figure(figsize=(7, 6))  
wedges, _, autotexts = plt.pie(
    sizes,
    labels=None, 
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    pctdistance=0.75,
    textprops={'fontsize': 10},
    wedgeprops={'edgecolor': 'white'}
)

plt.title('Total Amount by Gender', fontweight='bold', fontsize=14, pad=20)
plt.legend(wedges, labels, title='Gender', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
plt.axis('equal')
plt.tight_layout()
plt.show()


```

**Key Insight:** The majority of buyers are female, and their purchasing power surpasses that of their male counterparts.

<h2 style="color:green; font-size:28px;">🗺️ State :</h2>


```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

labels = sales_state['State']
sizes = sales_state['Orders']

colors = plt.cm.Set3.colors[:10]  

fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=None,  
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.75,
    textprops={'fontsize': 10},
    wedgeprops={'edgecolor': 'white'}
)

ax.set_title('Top 10 States by Orders', fontweight='bold', fontsize=14, pad=20)
ax.axis('equal')  
ax.legend(wedges, labels, title='States', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10, title_fontsize=11)

plt.tight_layout()
plt.show()
```

```python
#total sales from top 10 States.
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

labels = sales_state['State']
sizes = sales_state['Amount']

colors = plt.cm.tab10.colors  

plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    sizes,
    labels=None,  
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    pctdistance=0.75,
    textprops={'fontsize': 10},
    wedgeprops={'edgecolor': 'white'}
)

plt.title('Total Sales from Top 10 States', fontweight='bold', fontsize=16, pad=20)
plt.legend(wedges, labels, title='States', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
plt.axis('equal')  
plt.tight_layout()
plt.show()
```

**Key Insights :** From the visual analysis, it is evident that the majority of orders originate from Uttar Pradesh, while Maharashtra leads in total sales amount. Karnataka also demonstrates strong performance across both metrics.



<h2 style="color:green; font-size:28px;">🕊️ Marital Status :</h2>

```python
# plotting pie chart for distribution by marital status.

df['Marital_Status'] = df['Marital_Status'].map({0: 'Single', 1: 'Married'})

marital_counts = df['Marital_Status'].value_counts()

labels = marital_counts.index
sizes = marital_counts.values

colors = ['skyblue', 'lightcoral']  

plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    sizes,
    labels=None,  
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    pctdistance=0.75,
    textprops={'fontsize': 10},
    wedgeprops={'edgecolor': 'white'}
)

plt.title('Distribution by Marital Status', fontweight='bold', fontsize=14, pad=20)
plt.legend(wedges, labels, title='Marital Status', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#plotting gender wise amount pie chart.

df['Marital_Status'] = df['Marital_Status'].where(~df['Marital_Status'].isin([0, 1]), 
                                                   df['Marital_Status'].map({0: 'Single', 1: 'Married'}))
df['Gender'] = df['Gender'].replace({'M': 'Male', 'F': 'Female'}).fillna(df['Gender'])

sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum()
pivot = sales_state.pivot(index='Marital_Status', columns='Gender', values='Amount').fillna(0)

gender_labels = pivot.columns.tolist()
colors = ['skyblue', 'lightcoral', 'lightgreen', 'orange']  

for status in pivot.index:
    sizes = pivot.loc[status][gender_labels]
    if sizes.sum() == 0:
        continue  

    plt.figure(figsize=(6, 6))
    wedges, _, autotexts = plt.pie(
        sizes,
        labels=None,
        colors=colors[:len(sizes)],
        autopct='%1.1f%%',
        startangle=90,
        pctdistance=0.75,
        textprops={'fontsize': 10},
        wedgeprops={'edgecolor': 'white'}
    )
    plt.title(f'Gender-wise Amount ({status})', fontweight='bold', fontsize=14, pad=20)
    plt.legend(wedges, gender_labels, title='Gender', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()


```

**Key Insights:** From the visual analysis, it's evident that the dominant buyer demographic consists of **married women**, who demonstrate a notably high purchasing power. This trend highlights a valuable consumer segment for targeted marketing and product offerings.

<h2 style="color:#FFD700; font-size:28px;"> 📱 Product Category  :</h2>

```python
#Plotting a bar graph for Product category.

sns.set(rc={'figure.figsize': (20, 5)})

ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category', palette='muted')

plt.xticks(rotation=45)

for bars in ax.containers:
    ax.bar_label(bars)

plt.tight_layout()
plt.show()


```

```python
# Plotting pie chart for Top 10 Product Categories by Sales Amount.

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum()
sales_state = sales_state.sort_values(by='Amount', ascending=False).head(10)

labels = sales_state['Product_Category']
sizes = sales_state['Amount']

colors = [
    'lightcoral', 'skyblue', 'mediumseagreen', 'gold', 'orchid',
    'sandybrown', 'plum', 'lightsteelblue', 'khaki', 'pink'
]

plt.figure(figsize=(8, 8))
wedges, _, autotexts = plt.pie(
    sizes,
    labels=None,  
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.75,
    textprops={'fontsize': 10},
    wedgeprops={'edgecolor': 'white'}
)

plt.title('Top 10 Product Categories by Sales Amount', fontweight='bold', fontsize=14, pad=20)
plt.legend(wedges, labels, title='Product Category', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
